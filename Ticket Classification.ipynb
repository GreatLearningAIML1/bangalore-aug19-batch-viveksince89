```python
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
nltk.download('punkt')
nltk.download('stopwords')


def tokenzy(args):
  wordsy=[]
  for i in args:
    one=word_tokenize(i)
    stops=set(stopwords.words('english'))
    two=[x.translate(str.maketrans("","",string.punctuation)) for x in one if x.isalpha() and x not in stops]
    wordsy.append(two)
  return wordsy
```

```python
vocab=len(t.word_index)+1

import numpy as np
EMBEDDING_FILE = './glove.6B.100d.txt'

embeddings = {}
for o in open(EMBEDDING_FILE):
    word = o.split(" ")[0]
    g=word
    embd = o.split(" ")[1:]
    embd = np.asarray(embd, dtype='float32')
    p=embd
    embeddings[word] = embd

# create a weight matrix for words in training docs
embedding_matrix = np.zeros((vocab, 100))

for word, i in t.word_index.items():
	embedding_vector = embeddings.get(word)
	if embedding_vector is not None:
		embedding_matrix[i] = embedding_vector
```

```python
embedding=100
max_length=30
```

```python
from keras.models import Sequential
from keras.layers import Dropout,LSTM,Embedding,Flatten,Bidirectional,Dense
from keras import regularizers

model=Sequential()
model.add(Embedding(vocab,output_dim=embedding,weights=[embedding_matrix],input_length=max_length,trainable=True))
model.add(Bidirectional(LSTM(50,return_sequences=True)))
model.add(Flatten())
model.add(Dense(13,activity_regularizer=regularizers.l2(1e-2),activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau,EarlyStopping
reducedlr=ReduceLROnPlateau(monitor='loss',factor=0.2, patience=3,mode='auto',min_delta=0.0001)
es=EarlyStopping(monitor='loss', patience=3, mode='auto',min_delta=0.0001,baseline=None,restore_best_weights=False)
callbacks=[reducedlr,es]
```

```python
model.fit(X_train,y_train,callbacks=[reducedlr,es],epochs=25,batch_size=100)
```

```python
scores=model.evaluate(X_val,y_val,batch_size=100)
```

```python
from keras.models import Sequential
from keras.layers import LSTM,Dense,Embedding,Flatten,Dropout,Bidirectional
from keras import regularizers
from sklearn.model_selection import KFold

kfold=KFold(15,True,random_state=5)

inputs=np.concatenate((X_train,X_test),axis=0)
targetsy=np.concatenate((y_train,y_test),axis=0)

acc_per_fold=[]
loss_per_fold=[]
fold_no=1

for train, test in kfold.split(inputs,targetsy):
  model=Sequential()
  model.add(Embedding(vocab,output_dim=embedding,weights=[embedding_matrix],input_length=max_length,trainable=True))
  model.add(Bidirectional(LSTM(100,return_sequences=True)))
  #model.add(Dropout(0.2))
  #model.add(Bidirectional(LSTM(50,return_sequences=True)))
  model.add(Flatten())
  model.add(Dense(13,activity_regularizer=regularizers.l2(1e-2),activation='softmax'))
  model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
  model.fit(inputs[train],targetsy[train],callbacks=[reducedlr,es],epochs=10,batch_size=100)
  scores=model.evaluate(inputs[test],targetsy[test],batch_size=50)
    # Generate generalization metrics
  print(f'Score for fold {fold_no}: {model.metrics_names[0]} of {scores[0]}; {model.metrics_names[1]} of {scores[1]*100}%')
  acc_per_fold.append(scores[1] * 100)
  loss_per_fold.append(scores[0])
  acc_per_fold.clear()
  loss_per_fold.clear()

  # Increase fold number
  fold_no = fold_no + 1
  # == Provide average scores ==
  print('------------------------------------------------------------------------')
  print('Score per fold')
  for i in range(0, len(acc_per_fold)):
    print('------------------------------------------------------------------------')
    print(f'> Fold {i+1} - Loss: {loss_per_fold[i]} - Accuracy: {acc_per_fold[i]}%')
    print('------------------------------------------------------------------------')
    print('Average scores for all folds:')
    print(f'> Accuracy: {np.mean(acc_per_fold)} (+- {np.std(acc_per_fold)})')
    print(f'> Loss: {np.mean(loss_per_fold)}')
    print('-------------------------------------------------------------------------')
```
